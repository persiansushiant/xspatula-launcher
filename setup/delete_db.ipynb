# Delete Database

This notebook deletes configured database content.

Use it when you want to reset the project and start again.

## Warning

This is a destructive operation.

It can delete:

- Schemas
- Tables
- Process definitions
- Utility records
- Community records

Only run it when you really want to reset the database.

## Files used by default

This notebook runs from the `setup/` folder.

| Concept | Default file |
|---|---|
| Scheme | `./zzz/scheme_ai4sh_local_delete.json` |
| Pilot / Job | `job_delete_db.json` |

Most users only need to edit the Scheme file if database credentials change.

````python
from xspatula import Xpatula
````

## Create a Xpatula session

Because this notebook lives inside the `setup/` folder, `"."` means the current setup directory.

````python
xp = Xpatula(".")
````

## Simple mode

This is the recommended usage.

The library automatically uses the default delete Scheme and delete Pilot.

Because this notebook is interactive, Xspatula will ask for confirmation before deleting.

````python
xp.delete_database(interactive=True)
````

## What can be changed?

### Scheme

The Scheme controls which database is deleted and which PostgreSQL user is used.

Typical file:

```text
./zzz/scheme_ai4sh_local_delete.json
```

Edit this file when you need to change:

- PostgreSQL host
- PostgreSQL port
- Database name
- PostgreSQL superuser
- PostgreSQL password
- Delete behavior

### Pilot / Job

The Pilot / Job controls which delete process is executed.

Typical file:

```text
job_delete_db.json
```

Most users should not change this unless they are developing a custom reset flow.
