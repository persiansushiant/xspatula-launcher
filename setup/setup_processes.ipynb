## What this notebook does

This step can register:

- Root processes
- Process definitions
- Process parameters
- Process permissions
- Process dependencies

## Files used by default

This notebook runs from the `setup/` folder.

| Concept | Default file |
|---|---|
| Scheme | `./zzz/scheme_ai4sh_local_use.json` |
| Pilot / Job | `job_setup_processes.json` |

Most users only need to edit the Scheme file if the login user changes.

````python
xp = Xpatula(".")
````

## What can be changed?

### Scheme

The Scheme controls which Xspatula user is used to connect to the project database.

Typical file:

```text
./zzz/scheme_ai4sh_local_use.json
```

Edit this file when you need to change:

- Project user
- Project password
- Process execution flags
- Verbosity
- Overwrite behavior

### Pilot / Job

The Pilot / Job points to the setup process registration flow.

Typical file:

```text
job_setup_processes.json
```

Most users should not change this unless they are modifying the process registry itself.
