# Create Database

This notebook creates the PostgreSQL database and the core Xspatula database structure.

It uses the high-level `Xpatula` API so users do not need to call the old low-level functions directly.

## What this notebook does

This step can:

- Create the configured PostgreSQL database
- Create schemas
- Create tables
- Create database users and roles
- Generate project `.env` files

## Files used by default

This notebook runs from the `setup/` folder.

| Concept | Default file |
|---|---|
| Scheme | `./zzz/scheme_ai4sh_local_setup.json` |
| Pilot / Job | `job_setup_db.json` |

Most users only need to edit the Scheme file if database credentials change.

````python
from xspatula import Xpatula
````

## Create a Xpatula session

Because this notebook lives inside the `setup/` folder, `"."` means the current setup directory.

````python
xp = Xpatula(".")
````

## Simple mode

This is the recommended usage.

The library automatically uses the default setup Scheme and setup Pilot.

````python
xp.create_database(interactive=True)
````

## What can be changed?

### Scheme

The Scheme controls database settings and project behavior.

Typical file:

```text
./zzz/scheme_ai4sh_local_setup.json
```

Edit this file when you need to change:

- PostgreSQL host
- PostgreSQL port
- Database name
- PostgreSQL superuser
- PostgreSQL password
- Database users
- Process flags

### Pilot / Job

The Pilot / Job controls which setup process files are executed.

Typical file:

```text
job_setup_db.json
```

Most users should not change this unless they are developing a new setup flow.
